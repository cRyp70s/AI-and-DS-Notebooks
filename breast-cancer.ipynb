# One Hot Encode

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df
```

```python
ones = df["diagnosis_M"].sum()
zeros = df.shape[0]-ones.sum()
ones, zeros
```

Dataset is fairly balanced.

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns=["diagnosis_M"])
y = df["diagnosis_M"]
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
s_c = StandardScaler()
X=s_c.fit_transform(X)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

classifiers = {
    'LogisticReg': LogisticRegression(),
    'XGB': XGBClassifier(use_label_encoder=False, eval_metric="error"),
    'RandomForest': RandomForestClassifier(),
    }
```

```python
for clf_name in classifiers:
    print("Fitting", clf_name)
    classifiers[clf_name].fit(X_train, y_train)
    print(clf_name, "Score:", classifiers[clf_name].score(X_test, y_test))
    
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, classifiers["XGB"].predict(X_test)))
```

```python
classifiers["XGB"].save_model("breast_cancer_model.json")
```
